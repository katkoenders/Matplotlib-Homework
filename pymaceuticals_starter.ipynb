```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import os
import numpy as np
from scipy.stats import sem

# Study data files
Mouse_metadata = os.path.join("Resources","Mouse_metadata.csv")
Study_results = os.path.join("Resources","Study_results.csv")

# Read the mouse data and the study results
Mouse_metadata = pd.read_csv(Mouse_metadata)
Study_results = pd.read_csv(Study_results)
Mouse_metadata.head()

# Combine the data into a single dataset
Mouse_data = pd.merge(Mouse_metadata, Study_results, on = "Mouse ID")
Mouse_data.head()
```

## Summary statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
mean_of_tumor = Mouse_data.groupby(["Drug Regimen","Timepoint"]).mean()["Tumor Volume (mm3)"]
median_of_tumor = Mouse_data.groupby(["Drug Regimen","Timepoint"]).median()["Tumor Volume (mm3)"]
variance_of_tumor = Mouse_data.groupby(["Drug Regimen","Timepoint"]).var()["Tumor Volume (mm3)"]
std_of_tumor = Mouse_data.groupby(["Drug Regimen","Timepoint"]).std()["Tumor Volume (mm3)"]
sem_of_tumor = Mouse_data.groupby(["Drug Regimen","Timepoint"]).sem()["Tumor Volume (mm3)"]

summary_statistics = pd.DataFrame({"Mean":mean_of_tumor,
                                  "Median":median_of_tumor,
                                   "Vairance":variance_of_tumor,
                                    "Standard Deviation":std_of_tumor,
                                     "Standard Error":sem_of_tumor})

summary_statistics
```

## Bar plots

```python
# Generate a bar plot showing number of data points for each treatment regimen using pandas

bar_summary = summary_statistics[["Mean", "Median"]]
bar_summary = Mouse_data.groupby("Drug Regimen")["Drug Regimen"].value_counts()
bar_summary.plot(kind="bar", figsize = (20,5))

plt.title("Summary Statistics:\n"
          "Tumor Volume by Drug")
plt.ylabel("Tumor Volume (mm3)")
plt.tight_layout()
plt.show()


plt.savefig("summary_statistics.png")
```

## Line and scatter plots

```python
# Generate a line plot of time point versus tumor volume for a mouse treated with Capomulin
Capomulin_data = pd.DataFrame(pivot_Tumor_volume_mean["Capomulin"].reset_index())
Capomulin_data
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin 
x_axis = Capomulin_data["Timepoint"]
y_axis = Capomulin_data["Capomulin"]

plt.plot(x_axis,y_axis,marker='o',color='red')
plt.xlabel("Timepoint (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Capomulin Results :\n"
          "Tumor Volume Over Timespan")
plt.grid()
plt.show()
```

```python
Drug_table = Merge_data.loc[Merge_data["Drug Regimen"]=="Capomulin"]
Average = Drug_table.groupby(["Mouse ID"]).mean()
plt.scatter(Average["Weight (g)"],Average["Tumor Volume (mm3)"])
plt.xlabel("Weight (g)")
plt.ylabel("Tumor Volume (mm3)")
plt.show()
```

```python
Correlation = round(st.pearsonr(Average["Weight (g)"],Average["Tumor Volume (mm3)"])[0],2)
print(Correlation)
Model = st.linregress(Average["Weight (g)"],Average["Tumor Volume (mm3)"])
yvalues = Average["Weight (g)"]*Model[0]+Model[1]
plt.scatter(Average["Weight (g)"],Average["Tumor Volume (mm3)"])
plt.xlabel("Weight (g)")
plt.ylabel("Tumor Volume (mm3)")
plt.plot(Average["Weight (g)"],yvalues,color='green')
plt.show()
```
